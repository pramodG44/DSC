```python
import folium 
```

```python
spiderman_icon = folium.features.CustomIcon('spiderman.png', icon_size=(100,100))
m = folium.Map([19.102192, 73.005675], tips='', zoom_start=12)
folium.Marker([19.102192, 73.005675],
             tooltip='SPIDERMAN', 
             popup='<b>SPIDERMAN</b>',
             icon = spiderman_icon).add_to(m)

m
```

```python
scooter = folium.features.CustomIcon('scooter.png')
m = folium.Map([19.102192, 73.005675], tips='', zoom_start=12)
folium.CircleMarker([19.102366, 73.005760],
                    radius= 100, 
                    color = 'black', 
                    fill=True, 
                    fill_color = 'grey',
                    icon= scooter).add_to(m)
folium.Marker([19.102366, 73.005760],
             tooltip='SPIDERMAN', 
             popup='<b>SPIDERMAN</b>',
             icon = scooter).add_to(m)

m
```
